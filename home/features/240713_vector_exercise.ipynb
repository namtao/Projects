```python
# Độ dài của vecto
import numpy as np


def compute_vector_length(vecto):
    return np.sqrt(np.sum(vecto**2))
```

```python
# Phép tính tích vô hướng
def compute_dot_product(vecto1, vecto2):
    return np.dot(vecto1, vecto2)
```

```python
# Nhân vecto với ma trận
def matrix_multi_vector(matrix, vecto):
    return np.dot(matrix, vecto)
```

```python
# Nhân ma trận với ma trận
def matrix_multi_matrix(matrix1, matrix2):
    return np.dot(matrix1, matrix2)
```

```python
# Ma trận nghịch đảo
def inverse_matrix(matrix):
    return np.linalg.inv(matrix)
```

```python
# Eigenvector và eigenvalues
def compute_eigenvalues_eigenvectors(matrix):
    return np.linalg.eig(matrix)
```

```python
# Cosine Similarity
import numpy as np


def compute_cosine(A, B):
    dot_product = np.dot(A, B)
    norm_A = np.linalg.norm(A)
    norm_B = np.linalg.norm(B)
    cosine_sim = dot_product / (norm_A * norm_B)
    return cosine_sim
```
